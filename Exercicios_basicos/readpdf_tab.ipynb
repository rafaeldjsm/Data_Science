```python
import pandas as pd
from tabula import read_pdf
import seaborn as sns

df = read_pdf('RemuneraoZdosZConselheirosZDiretoresZeZEmpregadosZ-Z07-2021.pdf',pages='all')

# df = read_pdf('PLANILHA ORCAMENTARIA.pdf',pages='all')
```

```python
df0 = pd.DataFrame()

for k in df:
    df0 = pd.concat([df0,k])
df0
```

```python
df0.columns = df0.iloc[0,:]

df0 = df0[df0['UF'] != 'UF']
```

```python
df0.columns[5:]
```

```python
for k in df0.columns[5:]:
    df0[k] = df0[k].apply(lambda x: x.replace('.','').replace(',','.').replace('-',''))
    df0[k] = df0[k].astype(float)
```

```python
df0['TOTAL'] = df0['BÁSICA'] + df0['REMUNERAÇÕES']
```
