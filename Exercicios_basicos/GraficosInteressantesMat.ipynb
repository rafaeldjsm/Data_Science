# Verificando os padrões nos senos dos números inteiros.

```python
import matplotlib.pyplot as plt
from math import *

x = [*range(10000)]
y = [sin(k) for k in x]


fig = plt.figure(facecolor='white', figsize=(30, 7))

plt.scatter(x, y,  c="g",s=2)
plt.show()
```

# Espiral com aplicação de rotação de coordenadas

```python
xl = [0]
yl = [0.1]

for k in range(500):
  x = xl[-1]*cos(91*pi/180) + yl[-1]*sin(91*pi/180)
  y = -xl[-1]*sin(91*pi/180) + yl[-1]*cos(91*pi/180)
  xl.append(x*1.05)
  yl.append(y*1.05)


fig = plt.figure(facecolor='white', figsize=(10, 10))

plt.plot(xl, yl,  c="blue")
plt.show()
```

# Verificando os padrões das Ternas Pitagóricas.

```python
n = int(input("Até qual limite deseja ver as ternas pitagóricas: "))


a = []
b = []

for k in range(1,n):
  for j in range(1,k):
    bt = (k**2 - j**2)**0.5
    if bt == int(bt):
      a.append(j)
      b.append(bt)


fig = plt.figure(facecolor='white', figsize=(10, 10))

plt.scatter(a, b,  c="g",s=2)
plt.show()
```
