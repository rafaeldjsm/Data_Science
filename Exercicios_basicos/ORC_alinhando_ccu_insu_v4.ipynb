```python
ccu_list=[]

def PesquisaCCU(ccu_list,DFCI,col_c):

    sch_ccu = input("Qual o serviços deseja pesquisar?':").upper();
    
    df3=DFCI.copy()

    df3=df3['DESCRICAO DA COMPOSICAO'].str.contains(sch_ccu)
     #Procura na coluna "DESCRICAO DA COMPOSICAO"

    tp2=np.where(df3==1)

    df4=DFCI.iloc[tp2[0],:] 
    #Coluna com todas as composições que contenham o texto requerido

    pd.set_option('max_colwidth',800)
    pd.set_option('display.width', 1000)

    nl,nc=df4.shape

    if len(df4)==0:
        print('Serviço não encontrado na base de dados');
    else:
        print('Foram encontrados',nl,'serviços');

        df6=df4.copy()
        df6=df6.reset_index(drop=True)
        df6 = df6.style.set_properties(**{'text-align': 'left'})
        df6.hide_index().render()

        print ('\033[1m'+'\033[91m'+"Escolha na planilha de serviços e indique seu código")

        display(df6)

        ccu_esc = input("Dentre os serviços mostrados qual o código escolhido? ")

        ccu_list.append(ccu_esc)
       
    return (ccu_list)
```

```python
def expORC(L_q,L_c):

    df11=pd.DataFrame(L_q,L_c)
    df11.columns=['QTD']

    df12=pd.merge(df11, df5,left_index=True, right_index=True)

    df12['QTD TOTAL']=df12['COEFICIENTE']*df12['QTD']

    df13=df12.groupby(['CODIGO ITEM']).sum()

    df13=df12.groupby(['CODIGO ITEM']).agg({'QTD TOTAL':np.sum,'TIPO ITEM':'first','COEFICIENTE':'first'}).reset_index()

    dfin=df13.iloc[np.where(df13['TIPO ITEM']=='INSUMO')]

    dfcc=df13.iloc[np.where(df13['TIPO ITEM']=='COMPOSICAO')]

    lcod = dfcc['CODIGO ITEM'].values

    lqtd2 = dfcc['QTD TOTAL'].values

    return (dfin,lqtd2,lcod)

dfin,lqtd2,lcod=expORC(lqtd,ccu_list)

var_dict={}
var_dict[0]=lqtd2

var_dict2={}
var_dict2[0]=lcod

var_dict3={}
var_dict3[0]=dfin

i=1
while len(var_dict[i-1])>0:
    var_dict3[i],var_dict[i],var_dict2[i]=expORC(var_dict[i-1],var_dict2[i-1])
    i=i+1;
```
