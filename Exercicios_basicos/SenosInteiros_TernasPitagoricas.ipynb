# Verificando os padrões nos senos dos números inteiros.

```python
import plotly.express as px
from math import *

x = [*range(10000)]
y = [sin(k) for k in x]

fig = px.scatter(x=x,y=y)
fig.show()
```

```python
xl = [0]
yl = [0.1]

for k in range(500):
  x = xl[-1]*cos(91*pi/180) + yl[-1]*sin(91*pi/180)
  y = -xl[-1]*sin(91*pi/180) + yl[-1]*cos(91*pi/180)
  xl.append(x*1.05)
  yl.append(y*1.05)


fig = px.line(x=xl,y=yl)

fig.update_layout(
    width = 1000,
    height = 1000,
)

fig.show()
```

# Verificando os padrões das Ternas Pitagóricas.

```python
import plotly.graph_objects as go

n = int(input("Até qual limite deseja ver as ternas pitagóricas: "))


a = []
b = []

for k in range(1,n):
  for j in range(1,k):
    bt = (k**2-0 - j**2)**0.5
    if bt == int(bt):
      a.append(j)
      b.append(bt)


fig = px.scatter(x=a,y=b)

fig.update_layout(
    width = 1000,
    height = 1000,
)

fig.show()
```
