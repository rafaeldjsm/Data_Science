```python
import numpy as np
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro
from sklearn.preprocessing import StandardScaler
from datetime import timedelta

# Read the Excel file without specifying dtype initially
df = pd.read_excel('tai_games_24.xlsx').fillna(0)

# Extract numeric values from numeric columns
for col in df.columns[[2,3,4,5,8,10]]:
    df[col] = df[col].str.extract(r'pts\s*(\d+)').fillna(0).astype(int)

# Extract time values
for col in df.columns[[6,7,9]]:
    df[col] = pd.to_timedelta(df[col].str.extract(r'(\d{2}:\d{2}:\d{2})')[0]).dt.total_seconds().fillna(0)


# Display the modified dataframe
display(df)
```

```python
lin_team4 = df[df['COMPETIDOR'].str.contains("(4)",regex=False)]

df_time4 = lin_team4.index[0]

# Iterate through the columns you specified
for column_name in df.columns[2:-1]:

    team4_score = lin_team4[column_name].to_numpy()
    # Retirando os 0 zeros para evitar distorcao
    column_data = df[column_name][df[column_name] != 0]

    # Calculate the mean and median
    mean_score = np.mean(column_data)
    median_score = np.median(column_data)

    # Histogram
    plt.hist(column_data)#, bins=10)  # Adjust the number of bins as needed

    # Add a marker for the 4th team if available
    plt.axvline(team4_score, color='brown', linestyle='dashed', linewidth=2, label='Team 4')

    # Add a vertical line for the mean
    plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=2, label='Mean')

    # Add a vertical line for the median
    plt.axvline(median_score, color='green', linestyle='dashed', linewidth=2, label='Median')

    plt.xlabel(column_name)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column_name}')
    plt.legend()
    plt.show()

    # Calculate quartiles and other statistics
    q1 = np.percentile(column_data, 25)
    q2 = np.percentile(column_data, 50)  # Median (Q2)
    q3 = np.percentile(column_data, 75)
    iqr = q3 - q1

    # Create the boxplot and store the result in 'bp'
    bp = plt.boxplot(column_data, labels=[column_name])

    # Extract the whiskers positions
    whiskers = [item.get_ydata()[1] for item in bp['whiskers']]
    lower_whisker = whiskers[0] if whiskers else None
    upper_whisker = whiskers[1] if whiskers else None

    print(f"Lower Whisker Position: {lower_whisker}")
    print(f"Upper Whisker Position: {upper_whisker}")

    # Create the boxplot with annotations
    plt.boxplot(column_data, labels=[column_name])
    plt.ylabel(column_name)
    plt.title(f'Boxplot of {column_name} with Q1, Q2, Q3, and Limits')

    # Add annotations for Q1, Q2, and Q3
    plt.text(1.1, q1, f'Q1: {q1:.2f}', ha='left', va='center')
    plt.text(1.1, q2, f'Q2 (Median): {q2:.2f}', ha='left', va='center')
    plt.text(1.1, q3, f'Q3: {q3:.2f}', ha='left', va='center')

    plt.plot([1], [team4_score], 'ro', label='Team 4')  # 'ro' for red circle

    # Add annotations for the whiskers
    if lower_whisker is not None:
        plt.text(1.1, lower_whisker, f'Lower Whisker: {lower_whisker:.2f}', ha='left', va='bottom')
    if upper_whisker is not None:
        plt.text(1.1, upper_whisker, f'Upper Whisker: {upper_whisker:.2f}', ha='left', va='top')
    plt.legend()
    plt.show()

    # Create the swarmplot with annotations
    sns.swarmplot(y = column_data,
                hue=['Team 4' if k == df_time4 else '' for k in column_data.index])

    # Add a marker for the 4 team
    plt.axhline(team4_score, color='brown', linestyle='dashed', linewidth=2, label='Team 4')


    # Add a vertical line for the mean
    plt.axhline(mean_score, color='red', linestyle='dashed', linewidth=2, label='Mean')

    # Add a vertical line for the median
    plt.axhline(median_score, color='green', linestyle='dashed', linewidth=2, label='Median')
    plt.title(f'Swarmplot of {column_name}')
    plt.legend()
    plt.show()


    # Create a histogram with KDE plot
    # plt.figure(figsize=(10, 6))
    sns.histplot(column_data, kde=True)  # kde=True adds the KDE plot to the histogram
    plt.xlabel('Scores')
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {column_name} with KDE of Scores')

    # Add a marker for the Tem 4th
    plt.axvline(team4_score, color='brown', linestyle='dashed', linewidth=2, label='Team 4')


    # Add a vertical line for the mean
    plt.axvline(mean_score, color='red', linestyle='dashed', linewidth=2, label='Mean')

    # Add a vertical line for the median
    plt.axvline(median_score, color='green', linestyle='dashed', linewidth=2, label='Median')

    plt.legend()
    plt.show()

    # Shapiro-Wilk test for normality
    stat, p = shapiro(column_data)
    print(f"Shapiro-Wilk Test Statistic: {stat}")
    print(f"P-value: {p}")
    alpha = 0.05
    if p > alpha:
        print("Sample looks Gaussian (fail to reject H0)")
    else:
        print("Sample does not look Gaussian (reject H0)")
```

```python
# Create a new DataFrame 'df_stand' for standardized data
# Vou retirar a ultima linha (Time 55) para no interferir na padronizacao
df_stand = df.iloc[:-1,:].copy()

# Select the columns you want to standardize
columns_to_standardize = df.columns[2:-1]

# Iterate through the selected columns and perform standardization
for column in columns_to_standardize:
  if pd.api.types.is_numeric_dtype(df_stand[column]):  # Check if the column is numeric
    scaler = StandardScaler()
    df_stand[column] = scaler.fit_transform(df_stand[[column]])

# Display the standardized DataFrame
print("\nStandardized DataFrame:")
display(df_stand)
```

```python
# Plotando os histrogramas com kne do dataframe usando de todas as colunas, no mesmo grafico sobreposta e separadas por cor

# Create a figure and axes
plt.figure(figsize=(12, 8))

# Iterate through the columns and plot histograms with KDE on the same axes
for column in columns_to_standardize:
    sns.histplot(df_stand[column], kde=True, label=column, alpha=0.15)  # Use alpha for transparency

# Customize the plot
plt.xlabel('Standardized Values')
plt.ylabel('Frequency')
plt.title('Overlapping Histograms with KDE of Standardized Features')
plt.legend()
plt.show()
```

```python
df_time4_std = df_stand.iloc[df_time4,:]

# Create a figure and axes
plt.figure(figsize=(20, 8))

# Iterate through the columns and create violin plots side-by-side
for i, column in enumerate(columns_to_standardize):
  plt.subplot(1, len(columns_to_standardize), i + 1)  # Create subplots
  sns.violinplot(y=df_stand[column])

  # Add a horizontal line for Team 4 score
  time4_std = df_time4_std[column]
  plt.axhline(time4_std, color='red', linestyle='--', label=f'Team 4 {column}')
  plt.axhline(0, color='b', linestyle='dashed', linewidth=2, label='Mean')
  plt.axhline(df_stand.describe().iloc[5,i+1], color='k', linestyle='dashed', linewidth=2, label='Median')

  plt.title(column)
  plt.ylabel('Standardized Values')
  plt.legend()

# plt.tight_layout()  # Adjust subplot spacing
plt.show()
```

```python
# Create a figure and axes
plt.figure(figsize=(30, 14))
plt.rcParams.update({'font.size': 8})

# Iterate through the columns and create boxplots, swarmplots, and histograms with KDE side-by-side
for i, column in enumerate(columns_to_standardize):
    plt.subplot(2, len(columns_to_standardize), i + 1)  # Create subplots for boxplots
    sns.boxplot(y=df_stand[column], color='lightblue')  # Boxplot of original data
    sns.swarmplot(y=df_stand[column], color='red', size=3) # Swarmplot of standardized data

    # Add a horizontal line for Team 4 score
    time4_std = df_time4_std[column]
    plt.axhline(time4_std, color='b', linestyle='--', label=f'Team 4')

    # Add a vertical line for the mean
    plt.axhline(0, color='k', linestyle='dashed', linewidth=2, label='Mean')

    # Add a vertical line for the median
    plt.axhline(df_stand.describe().iloc[5,i+1], color='green', linestyle='dashed', linewidth=2, label='Median')

    plt.title(f'Boxplot and Swarmplot of {column}')
    plt.legend()

    plt.subplot(2, len(columns_to_standardize), i + 1 + len(columns_to_standardize))  # Create subplots for histograms
    sns.histplot(df_stand[column], kde=True, color='red', alpha=0.5)
    plt.axvline(time4_std, color='b', linestyle='--', label=f'Team 4')

    # Add a vertical line for the mean
    plt.axvline(0, color='k', linestyle='dashed', linewidth=2, label='Mean')

    # Add a vertical line for the median
    plt.axvline(df_stand.describe().iloc[5,i+1], color='green', linestyle='dashed', linewidth=2, label='Median')

    plt.title(f'Histogram and KDE of {column}')
    plt.legend()

plt.show()
```
